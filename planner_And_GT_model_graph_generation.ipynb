```python
def random_graph_planar_embedding(num_nodes):
    n=num_nodes
    i=0
    while True:
     G_er = nx.erdos_renyi_graph(n, 0.03)
     e=G_er.number_of_edges()
     k=(3*n-6)
     if e == k or e < k:
       break
     else:
         i+=1
         if i>10:
            print("not")
            return "can't generate"
    dict_=G_er.degree()
    remove_nodes=[]
    for i in range(n):
      if dict_[i]>6:
        remove_nodes.append(i)
    for i in range(n):
      if dict_[i]==0:
        remove_nodes.append(i)
    G_er.remove_nodes_from(remove_nodes)
    G_er=nx.convert_node_labels_to_integers(G_er)
    return G_er


```

```python
def GT_model_graph(l,t):
    l=l
    w=2*l+1
    G=nx.grid_2d_graph(w,w,periodic=False, create_using=None)
    nx.draw(G,with_labels=True)
    G_=nx.convert_node_labels_to_integers(G)
    x=[l,l*(w+1)-l,l*(w+1)+l,l*2*(w+1)-l]
    edge_conn=[(x[0],100*x[0]),(x[1],100*x[1]),(x[2],100*x[2]),(x[3],100*x[3])]
    edge_tree=[]
    for k in x:
      h=k*100
      for i in range(t):
          for j in range(i+1,i+3):
              edge_tree.append((h+i,h+i+j))
    edge_list=[*edge_tree,*edge_conn]
    G_.add_edges_from(edge_list)
    G_1=nx.convert_node_labels_to_integers(G_)
    return G_1
```

```python
l=1
t=10
num_of_nodes=(2*l+1)**2 + t*3*4
G=GT_model_graph(l,t)
```
